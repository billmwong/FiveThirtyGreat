NBA stats recently took down their shot logs for each player. For statistics like nearest defender, number of dribbles, and touch time, NBA stats only has the averages for each player. This isn't particularily useful when we are trying to make a shot predictive model. So, we must scrape for this data through a site called nba savant, which has this data from 2014-2015 season. 

```python
import urllib
import urllib2
from bs4 import BeautifulSoup
import pandas as pd
```

```python
id='201939' # Stephen Curry with the shots, boy
```

```python
url = 'http://nbasavant.com/ajax/getShotsByPlayer.php?hfST=&hfQ=&pid%5B%5D='+id+'&hfSZB=&hfSZA'+ \
'=&hfSZR=&ddlYear=2014&txtGameDateGT=&txtGameDateLT=&ddlGameTimeGT_min=&ddlGameTimeGT_sec='+ \
'&ddlGameTimeLT_min=&ddlGameTimeLT_sec=&ddlShotClockGT=&ddlShotClockLT=&ddlDefDistanceGT=&ddlDefDistanceLT'+ \
'=&ddlDribblesGT=&ddlDribblesLT=&ddlTouchTimeGT=&ddlTouchTimeLT=&ddlShotDistanceGT=&ddlShotDistanceLT'+ \
'=&ddlTeamShooting=&ddlTeamDefense=&hfPT=&ddlGroupBy=player&ddlOrderBy=shots_made_desc&hfGT=0%7C&ddlShotMade'+ \
'=&ddlMin=0&player_id='+id+'&data=null&_=1459882092105'

print url
```

```python
htmlfile=urllib.urlopen(url)
html=htmlfile.read()
print html

# soup = BeautifulSoup(html)
# print soup
```
