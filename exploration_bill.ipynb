```python
# steph = get_player('Stephen', 'Curry', just_id=False)
# player.PlayerShotTracking(steph['PERSON_ID']).shot_clock_shooting()

def getShotClockSplits(playerString):
    splitString = playerString.split(' ')
    thisPlayer = get_player(splitString[0], splitString[1], just_id=False)
    df = player.PlayerShotTracking(thisPlayer['PERSON_ID']).shot_clock_shooting()
    
    N = 8
    menMeans = df['FG_PCT']

    ind = np.arange(N)  # the x locations for the groups
    width = 0.5       # the width of the bars

    fig, ax = plt.subplots()
    rects1 = ax.bar(ind, menMeans, width, color='r')

    # add some text for labels, title and axes ticks
    ax.set_ylabel('FG%')
    ax.set_title('FG% by Shot Clock Range for '+playerString)
    ax.set_xticks(ind + (width)/2)
    ax.set_xticklabels(df['SHOT_CLOCK_RANGE'])

    def autolabel(rects):
        # attach some text labels
        for rect in rects:
            height = rect.get_height()
            ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                    '%d' % int(height),
                    ha='center', va='bottom')

#     autolabel(rects1)

    plt.show()
    return df

getShotClockSplits('Stephen Curry')
```
